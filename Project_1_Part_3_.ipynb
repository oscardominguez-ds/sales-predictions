```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive')

import seaborn as sns
import numpy as np

filename = '/content/sales_predictions'
df = pd.read_csv(filename)
```

```python
#Our histograms tells us that there are more units sold around the 2,000 rupees price than any other price. 

df['Item_Outlet_Sales'].hist(bins = 'auto');
plt.xticks(rotation = 45)
plt.ticklabel_format(useOffset=False, style='plain')
```

```python
#I used the same column from above to see the same data in a different way
#This boxplot tells us the outliers. Outliers are the stores that generated more sales than the average.
#Item_Outlet_Sales boxplot tells give us the average money in sales is 2,000 rupees.


fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (8,5))
boxplots = axes.boxplot([df['Item_Outlet_Sales']],
           notch = True,
           labels=['Item_Outlet_Sales'],
           widths = .7,
           patch_artist=True,
           medianprops = dict(linestyle='-', linewidth=2, color='Yellow'),
           boxprops = dict(linestyle='--', linewidth=2, color='Black', facecolor = 'blue', alpha = .4)
          );
```

```python
#This boxplot tells us that the median retail price is 150 for items sold with no outliers.

fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (8,5))
boxplots = axes.boxplot([df['Item_MRP']],
           notch = True,
           labels=['Item_MRP'],
           widths = .7,
           patch_artist=True,
           medianprops = dict(linestyle='-', linewidth=2, color='Yellow'),
           boxprops = dict(linestyle='--', linewidth=2, color='Black', facecolor = 'blue', alpha = .4)
          );
```

```python
#Heatmap shows that there is a moderate correlation between Item_Outlet_Sales and Item_MRP
corr = df.corr()
sns.heatmap(corr, cmap = 'Greens', annot = True)
```
